# Calculate correlation matrix & Create heatmap for 2020

```python
report_2020_selected  = report_2020_sorted[['RANK', 'Happiness score','Explained by: GDP per capita', 'Explained by: Social support',
       'Explained by: Healthy life expectancy',
       'Explained by: Freedom to make life choices',
       'Explained by: Generosity', 'Explained by: Perceptions of corruption' ]]
```

```python
# Calculate correlation matrix
correlation_matrix_2020 = report_2020_selected.corr()
```

```python
# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_2020, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap for 2020')
plt.show()
```

# Create a categorical plot and interpret the results.

```python
report_2020_sorted.loc[report_2020_sorted['Happiness score'] < 4 , 'Happiness Category'] = 'Low'
```

```python
report_2020_sorted.loc[(report_2020_sorted['Happiness score'] >= 4) & (report_2020_sorted['Happiness score'] < 6), 'Happiness Category'] = 'Medium'
```

```python
report_2020_sorted.loc[report_2020_sorted['Happiness score'] >= 6 , 'Happiness Category'] = 'High'
```

```python
report_2020_sorted['Happiness Category'].value_counts(dropna = False)
```

```python
sns.catplot(x='Explained by: GDP per capita', y='Happiness score', hue="Happiness Category", data=report_2020_sorted)

# Add labels and title
plt.xlabel('Explained by: GDP per capita')
plt.ylabel('Happiness score')
plt.title('Happiness Category Plot for 2020')
```

categorical plot indicating "higher GDP leads to higher happiness" suggests that countries with a higher GDP per capita tend to have happier populations.
